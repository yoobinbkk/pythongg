## 결측치 데이타 처리

- 결측치는 데이터 분석 결과에 엄청난 영향을 미치기에 존재여부 확인부터 중요

- None를 포함하며, 결측값은 ‘NaN’로 표시


- 결측값 확인 메소드 
    * isnull()
    * notnull()
    
    
- 결측값 처리 	
    (1) 제거 : dropna()      	
    (2) 대치 : fillna()


```python
import pandas as pd
import numpy as np 

df = pd.DataFrame(np.ones((4,4)),index=['A','B','C','D'])
df
```

```python
# 결측치 만들기 ( NaN, nan)
df.loc['B', 1] = np.NaN
df

df.loc['B':, 2] = np.NaN
df
```

```python
# 누적합을 구할 때
df.cumsum()
```

```python
# 결측치 확인
print(df.isnull())
print(df.notnull().sum())
```

```python
# 결측치 해결 -> 대체
# df.dropna()
df.fillna(0)
```

```python
df.fillna('Unknown')
```

```python
df.fillna({1:'NULL', 2:-1})
```
